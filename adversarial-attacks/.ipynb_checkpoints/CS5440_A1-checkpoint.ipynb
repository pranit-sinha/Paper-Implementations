```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable
import copy
import numpy as np
import matplotlib.pyplot as plt
import dill
import random
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
def zero_gradients(x):
    if isinstance(x, torch.Tensor):
        if x.grad is not None:
            x.grad.detach_()
            x.grad.zero_()
    elif isinstance(x, collections.abc.Iterable):
        for elem in x:
            zero_gradients(elem)
```

```python
class SmallAlexNet(nn.Module):
    def __init__(self, num_classes=10):
        super(SmallAlexNet, self).__init__()
        
        self.conv1 = nn.Conv2d(3, 384, kernel_size=5, padding=2)
        self.pool1 = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
        
        self.conv2 = nn.Conv2d(384, 192, kernel_size=5, padding=2)
        self.pool2 = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
        
        self.flat_size = 192 * 8 * 8
        
        self.fc1 = nn.Linear(self.flat_size, 384)
        self.fc2 = nn.Linear(384, 192)
        self.fc3 = nn.Linear(192, num_classes)
        self.relu = nn.ReLU()
    
    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = self.pool1(x)
        x = F.local_response_norm(x, size=5, alpha=1e-4, beta=0.75, k=2)
        
        x = self.conv2(x)
        x = self.relu(x)
        x = self.pool2(x)
        x = F.local_response_norm(x, size=5, alpha=1e-4, beta=0.75, k=2)
        
        x = x.view(-1, self.flat_size)
        
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        
        return x
```

```python
model = SmallAlexNet()
model.load_state_dict(torch.load('model_weight.pt', weights_only=True))
model.eval()

labels = unpickle('cifar-10-batches-py/batches.meta')[b'label_names']
labels = [label.decode("utf-8") for label in labels]
classes = tuple(labels)
test_data = unpickle('cifar-10-batches-py/test_batch')
test_images = test_data[b'data'].reshape(10000, 3, 32, 32).astype(np.float32) / 255.0
test_images = torch.tensor(test_images).to(torch.float32)
test_labels = np.array(test_data[b'labels'])
test_labels = torch.tensor(test_labels).to(torch.float32)

# referred to https://stackoverflow.com/a/68123869 since values well known for CIFAR 10
mean = [0.49139968, 0.48215827 ,0.44653124]
std = [0.24703233, 0.24348505, 0.26158768]
test_images = (test_images - np.array(mean).reshape(1, 3, 1, 1)) / np.array(std).reshape(1, 3, 1, 1)
                                                                                                                                                                                     
testset = torch.utils.data.TensorDataset(test_images, test_labels)
testloader = torch.utils.data.DataLoader(testset, batch_size = 64, shuffle = False, num_workers = 2)
```

```python
from sklearn.metrics import classification_report

y_true = np.empty(10000, dtype=int)
y_pred = np.empty(10000, dtype=int)
idx = 0

with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = model(images)
        # outputs are class-wise energies, so idx of highest energy -> predicted class
        _, preds = torch.max(outputs, 1)
        y_true[idx:idx+64] = labels.numpy()
        y_pred[idx:idx+64] = preds.numpy()
        idx += 64

print(classification_report(y_true, y_pred))
```

```python
untrained = SmallAlexNet()
untrained.eval()

y_true = np.empty(10000, dtype=int)
y_pred = np.empty(10000, dtype=int)
idx = 0

with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = untrained(images)
        _, preds = torch.max(outputs, 1)
        y_true[idx:idx+64] = labels.numpy()
        y_pred[idx:idx+64] = preds.numpy()
        idx += 64

print(classification_report(y_true, y_pred))
```

Note from the warnings above that there are three classes the model never predicts

```python
# cf. section 3.1 shafahi et al. we don't explicitly tune beta because the perturbations with this value seemed invisible when checked
beta = 0.25 * (192**2) / (3 * 32 * 32) # = 3
# ended up trying lower beta for speed
```

```python
# cf. algorithm 1, shafahi et al.

def poisonFrogs(model, b, t, beta, lr=0.01, maxIters=1000):
    b = b.clone().requires_grad_(False)
    t = t.clone().requires_grad_(False)
    x = b.clone().requires_grad_(True)
    
    for _ in range(maxIters):
        L_p = torch.linalg.vector_norm(model(x) - model(t))**2
        x_hat = x - lr * torch.autograd.grad(L_p, x)[0]
        x = (x_hat + lr * beta * b) / (1 + beta * lr)
        x.data = torch.clamp(x, 0, 1)
    
    return x.detach()
```

Trying attack on random images:

```python
plt.figure(figsize=(20, 50))
for i, tclass in enumerate(classes):
    tidx = (test_labels == i).nonzero()[random.randrange(0,1000)].item()
    bidx = (test_labels != i).nonzero()[random.randrange(0,1000)].item()
    
    b = test_images[bidx].to(torch.float32)
    t = test_images[tidx].to(torch.float32)
    
    p = poisonFrogs(model, b.view(1, *test_images.shape[1:]), t.view(1, *test_images.shape[1:]), beta).squeeze()
    with torch.no_grad():
            p_out = model(p)
            _, p_pred = torch.max(p_out, 1)
            
            b_out = model(b)
            _, b_pred = torch.max(b_out, 1)

    # code for visualization written with assistance from deepseek
    plt.subplot(10, 3, i*3 + 1)
    pdisplay = p * torch.tensor(std).view(3, 1, 1) + torch.tensor(mean).view(3, 1, 1)
    pdisplay = np.clip(pdisplay.permute(1, 2, 0).cpu().numpy(), 0, 1)
    plt.imshow(pdisplay)
    plt.title(f"Poisoned pred: {classes[p_pred.item()]}", fontsize=10)
    plt.axis('off')


plt.tight_layout(pad=1.0)
plt.show()
```

Attempting to find successful attack instances:

```python
p_images = []
b_images = []
t_classes = []

for tclass, _ in enumerate(classes):
    trial = (test_labels != tclass).nonzero(as_tuple=True)[0]
    idxes = (test_labels == tclass).nonzero(as_tuple=True)[0][0].item()
    
    flag = False

    for i in trial:
        i = i.item()
        b = test_images[i].view(1, *test_images.shape[1:]).to(torch.float32)
        t = test_images[tidx].view(1, *test_images.shape[1:]).to(torch.float32)
    
        x = poison_frogs(model, b, t, beta)

        with torch.no_grad():
            p_out = model(x)
            _, p_pred = torch.max(p_out, 1)
            
            if p_pred.item() == tclass:
                flag = True
                p_images.append(x.squeeze().to(torch.float32))
                b_images.append(b.squeeze().to(torch.float32))
                t_classes.append(tclass)
                break
    
    if not flag:
        print("Attack failed for ", tclass)
```

```python
def deepFool(image, net, overshoot=0.02, max_iter=50):

    if image.dim() == 3:
        image = image.unsqueeze(0)

    f_image = net.forward(Variable(image, requires_grad=True)).data.cpu().numpy().flatten()
    I = (np.array(f_image)).flatten().argsort()[::-1]
    I = I[0:10]
    label = I[0]

    inshape = image.shape[1:]

    poisoned = copy.deepcopy(image)
    w = np.zeros(inshape)
    sum_r = np.zeros(inshape)
    i = 0

    x = Variable(poisoned, requires_grad=True)
    fs = net.forward(x)
    k_i = label

    while k_i == label and i < max_iter:
        pert = np.inf
        fs[0, I[0]].backward(retain_graph=True)
        del_f_k_hat = x.grad.data.cpu().numpy().copy()

        for k in range(10):
            zero_gradients(x)
            fs[0, I[k]].backward(retain_graph=True)
            del_f_k = x.grad.data.cpu().numpy().copy()

            w_k = del_f_k - del_f_k_hat
            f_k = (fs[0, I[k]] - fs[0, I[0]]).data.cpu().numpy()
            pert_k = abs(f_k) / np.linalg.norm(w_k.flatten())
    
            if pert_k < pert:
                pert = pert_k
                w = w_k

        r_i = (pert + 1e-4) * w / np.linalg.norm(w)
        sum_r = np.float32(sum_r + r_i)

        poisoned = image + (1 + overshoot) * torch.from_numpy(sum_r)

        x = Variable(poisoned, requires_grad=True)
        fs = net.forward(x)
        k_i = np.argmax(fs.data.cpu().numpy().flatten())

        i += 1

    sum_r = (1+overshoot)*sum_r

    return sum_r, poisoned
```

```python
plt.figure(figsize=(20, 10))

res = []

for i, name in enumerate(classes):
    idx = (test_labels == i).nonzero()[random.randrange(0,1000)].item()

    og = test_images[idx].to(torch.float32)
    
    perturbation, pert = deepFool(og, model)
    
    with torch.no_grad():
        output_og = model(og.unsqueeze(0))
        _, pred_og = torch.max(output_og, 1)
        
        output_pert = model(pert.squeeze(0))
        _, pred_pert = torch.max(output_pert, 1)
    
    res.append({
        'Class': name,
        'Original Prediction': classes[pred_og.item()],
        'Perturbed Prediction': classes[pred_pert.item()],
        'Perturbation': np.linalg.norm(perturbation)
    })
    
    plt.subplot(2, 10, i + 1)
    og_unnorm = og * torch.tensor(std).view(3, 1, 1) + torch.tensor(mean).view(3, 1, 1)
    og_unnorm = og_unnorm.permute(1, 2, 0).cpu().numpy()
    og_unnorm = np.clip(og_unnorm, 0, 1)
    plt.imshow(og_unnorm)
    plt.title(f"Original pred: {classes[pred_og.item()]}")
    plt.axis('off')

    plt.subplot(2, 10, i + 11)
    pert_unnorm = pert.squeeze()
    pert_unnorm = pert_unnorm * torch.tensor(std).view(3, 1, 1) + torch.tensor(mean).view(3, 1, 1)
    pert_unnorm = pert_unnorm.permute(1, 2, 0).cpu().numpy()
    pert_unnorm = np.clip(pert_unnorm, 0, 1)
    plt.imshow(pert_unnorm)
    plt.title(f"Perturbed pred: {classes[pred_pert.item()]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))

res = []

for i, name in enumerate(classes):
    indices = (test_labels == i).nonzero().flatten()
    indices = indices[torch.randperm(len(indices))[:100]]
    mindist = float('inf')
    minimg = None
    for j in indices:
        og = test_images[j].to(torch.float32)
        dist, pert = deepFool(og, model, overshoot=0.0002)

        if np.linalg.norm(dist) < mindist :
            mindist = np.linalg.norm(dist)
            minimg = pert.squeeze(0)
    
            with torch.no_grad():
                output_og = model(og.unsqueeze(0))
                _, pred_og = torch.max(output_og, 1)
                
                output_pert = model(minimg)
                _, pred_pert = torch.max(output_pert, 1)

    res.append({
        'Class': name,
        'Original Prediction': classes[pred_og.item()],
        'Perturbed Prediction': classes[pred_pert.item()],
        'Perturbation': mindist
    })
    
    plt.subplot(2, 10, i + 1)
    og_unnorm = og * torch.tensor(std).view(3, 1, 1) + torch.tensor(mean).view(3, 1, 1)
    og_unnorm = og_unnorm.permute(1, 2, 0).cpu().numpy()
    og_unnorm = np.clip(og_unnorm, 0, 1)
    plt.imshow(og_unnorm)
    plt.title(f"Original pred: {classes[pred_og.item()]}")
    plt.axis('off')

    # Display perturbed image
    plt.subplot(2, 10, i + 11)
    pert_unnorm = pert.squeeze()
    pert_unnorm = pert_unnorm * torch.tensor(std).view(3, 1, 1) + torch.tensor(mean).view(3, 1, 1)
    pert_unnorm = pert_unnorm.permute(1, 2, 0).cpu().numpy()
    pert_unnorm = np.clip(pert_unnorm, 0, 1)
    plt.imshow(pert_unnorm)
    plt.title(f"Perturbed pred: {classes[pred_pert.item()]}")
    plt.axis('off')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
